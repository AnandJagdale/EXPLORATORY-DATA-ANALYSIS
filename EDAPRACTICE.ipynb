OUTLIERS & MISSING VALUES HANDLING BY IMPUTATION TECHNIQUES
a) Numerical feature or data= use Mean & Median imputation technique
b) Categorical feature or data= use Mode imputation techniques
NOTE:- Imputation techniques is same for handling both outliers & missing values

```python
#Cal. mean for Ozone feature
mean= data_cleaned3['Ozone'].mean()
print(mean) 
```

```python
#Mean imputation for fill null values of Ozone feature
data_cleaned3['Ozone']=data_cleaned3['Ozone'].fillna(mean) 
```

```python
#Check
data_cleaned3['Ozone'].isnull().sum() 
```

```python
#Mode imputation for fill null values of Weather feature(ddtype= object)
obj_columns=data_cleaned3[['Weather']]
obj_columns= obj_columns.fillna(obj_columns.mode().iloc[0])
obj_columns.isnull().sum() 
```

```python
#Mean imputation for fill null values of Solar feature(ddtype= float64)
data_cleaned3['Solar']= data_cleaned3['Solar'].fillna(mean) 
```

```python
#Drop unwanted column [Weather]
data_cleaned3= data_cleaned3.drop('Weather',axis=1)
data_cleaned3 
```

```python
#Combined a (obj_columns) with dataset by using concat function
data_cleaned3=pd.concat([data_cleaned3,obj_columns],axis=1)
data_cleaned3.isnull().sum() 
```

```python
#Create dummy data for column Weather
data_cleaned4=pd.get_dummies(data,columns=['Weather'])
data_cleaned4
```

```python
#Drop the any null values which is present in dataset
data_cleaned4=data_cleaned4.dropna()
data_cleaned4 
```

```python
#In Month feature column having 23rd obs.(May) repalce it by (5)
data_cleaned4['Month']= data_cleaned4['Month'].replace(['May'],'5') 
data_cleaned4 
```

```python
#Normalized the data
from numpy import set_printoptions
from sklearn.preprocessing import MinMaxScaler
```

```python
array= data_cleaned4.values
scaler= MinMaxScaler(feature_range=(0,1))
rescaledX= scaler.fit_transform(array[:,0:5])
```

```python
#Summarize the normalized data
set_printoptions(precision=2)
print(rescaledX[0:5,:]) 
```

```python
#Standardized the data
from sklearn.preprocessing import StandardScaler
```

```python
array= data_cleaned4.values
scaler= StandardScaler().fit(array)
rescaledX= scaler.transform(array) 
```

```python
#Summarize the standardized data
set_printoptions(precision=2)
print(rescaledX[0:5,:]) 
```
